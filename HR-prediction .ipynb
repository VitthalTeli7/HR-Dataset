```python
def empirical_analysis(col):
    mu=hr_df[col].mean()
    sigma=hr_df[col].std()
    minus1 = mu - sigma ; plus1 = mu + sigma
    minus2 = mu - (2*sigma) ; plus2 = mu + (2*sigma)
    minus3 = mu - (3*sigma) ; plus3 = mu + (3*sigma)
    x1 = hr_df[(hr_df[col] > minus1) & (hr_df[col] < plus1)]
    x2 = hr_df[(hr_df[col] > minus2) & (hr_df[col] < plus2)]
    x3 = hr_df[(hr_df[col] > minus3) & (hr_df[col] < plus3)]
    return(x1,x2,x3)
```

```python
for i in num: 
    x=empirical_analysis(i)
    print(x)
```
